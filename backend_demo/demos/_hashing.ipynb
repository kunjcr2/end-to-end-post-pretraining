```python
import jwt
import datetime
from datetime import timedelta

SECRET = "cqiuwriwoqnfurgvnfoirvnioqwngqiwrjgvbqwohrngfckwrungkgb"
ALGORITHM = "HS256"
EXPIRE_MIN = 30

def encode_jwt(payload: dict):
    payload_2 = payload.copy()
    td_obj = timedelta(minutes=EXPIRE_MIN)
    encoded_jwt = jwt.encode(
        payload_2, 
        SECRET, 
        algorithm=ALGORITHM
    )
    return encoded_jwt

def decode_jwt(token: str):

    decoded_jwt = jwt.decode(token, SECRET, algorithms=[ALGORITHM])
    return decoded_jwt
```

```python
jwt_token = encode_jwt({
    "id": 4,
    "email": "hello123@gmail.com",
    "exp": datetime.datetime.utcnow() + timedelta(minutes=EXPIRE_MIN)
})
```

```python
decode_jwt(
    jwt_token
)
```
